#### Package constants

```python
# Vocabulary domain - earth system.
DOMAIN_EARTH_SYSTEM = u'earth-system'

# Set of allowed vocabulary domains.
DOMAIN_SET = set([
    DOMAIN_EARTH_SYSTEM
])

# Vocabulary sub-domain - modelling.
SUBDOMAIN_MODELLING = u'modelling'

# Vocabulary sub-domain - observations.
SUBDOMAIN_OBSERVATIONS = u'observations'

# Set of allowed vocabulary sub-domains.
SUBDOMAIN_SET = set([
    SUBDOMAIN_MODELLING,
    SUBDOMAIN_OBSERVATIONS
])

# Governance state - the term is pending review.
GOVERNANCE_STATUS_PENDING = 'pending'

# Governance state - the term is accepted.
GOVERNANCE_STATUS_ACCEPTED = 'accepted'

# Governance state - the term is rejected.
GOVERNANCE_STATUS_REJECTED = 'rejected'

# Governance state - the term is obsolete.
GOVERNANCE_STATUS_OBSOLETE = 'obsolete'

# Set of allowed governance states.
GOVERNANCE_STATUS_SET = set([
    GOVERNANCE_STATUS_PENDING,
    GOVERNANCE_STATUS_ACCEPTED,
    GOVERNANCE_STATUS_REJECTED,
    GOVERNANCE_STATUS_OBSOLETE
])
```

#### Package defaults.

```python
# Default domain.
DEFAULT_DOMAIN = DOMAIN_EARTH_SYSTEM

# Default subdomain.
DEFAULT_SUBDOMAIN = SUBDOMAIN_MODELLING

# Default governance state.
DEFAULT_GOVERNANCE_STATUS = GOVERNANCE_STATUS_PENDING
```

#### Package type system.

```python
class Superset(object):
    """The superset of all managed vocabularies.
    
    """
    def __init__(self, termsets=None):
        """Object constructor.
        
        """
        self.termsets = termsets or []


    def __repr__(self):
        """Object representation.
        
        """
        return u"Superset --> {} termsets".format(len(self.termsets))

    
    @property
    def domains(self):
        """Gets set of unique domains declared across all termsets within superset.
        
        """ 
        pass
    
    
    def get_termset(self, term_type):
        for termset in self.termsets:
            if termset.typeof == term_type:
                return termset
    
    
class ScopedSet(object):
    """A scoped set of managed vocabularies.
    
    """
    def __init__(self, typeof, termsets=None, uid=None):
        """Object constructor.
        
        """
        self.typeof = unicode(typeof).strip()
        self.termsets = termsets or []
        self.uid = unicode(uid or (uuid.uuid4()))


    def __repr__(self):
        """Object representation.
        
        """
        return u"ScopedSet {0} --> {1} termsets".format(self.typeof, len(self.termsets))

    
class Termset(object):
    """A managed collection of vocabulary terms.
    
    """
    def __init__(self, domain=None, subdomain=None, typeof=None, subtype=None, terms=None, uid=None):
        """Object constructor.
        
        """
        self.domain = unicode(domain or DEFAULT_DOMAIN)
        self.subdomain = unicode(subdomain or DEFAULT_SUBDOMAIN)
        self.subtype = unicode(subtype).strip()
        self.terms = terms or []
        self.typeof = unicode(typeof).strip()
        self.uid = unicode(uid or (uuid.uuid4()))

        
    def __repr__(self):
        """Object representation.
        
        """
        return u"{0}.{1}.{2}.{3} --> {4}".format(self.domain, self.subdomain, self.typeof, self.subtype, len(self.terms))


    @staticmethod
    def create(namespace, uid):
        """Type instance factory method.
        
        """
        namespace = namespace.split(".")
        if len(namespace) != 4:
            raise ValueError("Termset namespace is invalid.")
        namespace = [i.strip() for i in namespace]
        
        return Termset(domain=namespace[0],
                       subdomain=namespace[1],
                       typeof=namespace[2],
                       subtype=namespace[3],
                       uid=uid)


    def validate(self):
        """Validates a term.
        
        """
        result = []
        for term in self.terms:
            result += term.validate()
            
        return result
    
    
    def is_valid(self):
        """Returns a flag indicating whether the term is deemed to be valid.
        
        """
        return len(self.validate()) == 0
        

class Term(object):
    """A vocabulary term.
    
    """
    def __init__(self, name, uid=None, create_date=None, status=None):
        """Object constructor.
        
        """
        self.name = unicode(name).strip()
        self.create_date = create_date or datetime.datetime.utcnow()
        self.status = unicode(status or DEFAULT_GOVERNANCE_STATUS)
        self.uid = unicode(uid or (uuid.uuid4()))
        
        
    def __repr__(self):
        """Object representation.
        
        """
        return u"{0} --> {1}".format(self.name, self.status)
    
    
    @staticmethod
    def create(name):
        """Type instance factory method.
        
        """
        return Term(name)
    

    def _validate_name(self):
        """Validates a term's name.
        
        """
        if self.name is None:
            raise ValueError('Term name is undefined')
        if not isinstance(self.name, unicode):
            raise ValueError('Term name is not a unicode type')
        if not len(self.name.strip()):
            raise ValueError('Term name is zero length')
            
            
    def _validate_create_date(self):
        """Validates a term's creation date.
        
        """
        if self.create_date is None:
            raise ValueError('Term create date is undefined')
        if not isinstance(self.create_date, datetime.datetime):
            raise ValueError('Term create date is not a datetime.datetime type')
            
            
    def _validate_status(self):
        """Validates a term's governance status.
        
        """
        if self.status not in GOVERNANCE_STATUS_SET:
            raise ValueError('Term governance status is invalid.  Valid statae = {}'.format(GOVERNANCE_STATUS_SET))

    
    def _validate_uid(self):
        """Validates a term's unique identifier.
        
        """
        if self.uid is None:
            raise ValueError('Term unique identifier is undefined')
        if not isinstance(self.name, unicode):
            raise ValueError('Term unique identifier is not a unicode type')
        try:
            uuid.UUID(self.uid)
        except ValueError:
            raise ValueError('Term unique identifier is not a valid UUID')            
            
            
    def validate(self):
        """Validates a term.
        
        """
        result = []
        for parser in (
            self._validate_create_date,
            self._validate_name,
            self._validate_status,
            self._validate_uid
            ):
            try:
                parser()
            except ValueError as err:
                result.append(err)
        
        return result
    
    
    def is_valid(self):
        """Returns a flag indicating whether the term is deemed to be valid.
        
        """
        return len(self.validate()) == 0
```
