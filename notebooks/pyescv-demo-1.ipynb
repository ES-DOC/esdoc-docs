## Governance

```python
pyescv.constants.GOVERNANCE_STATUS_SET
```

```python
model = ipsl_models['ipsl-cm5a-lr']
pyescv.reject(model)
assert model.status == pyescv.constants.GOVERNANCE_STATUS_REJECTED
pyescv.deprecate(model)
assert model.status == pyescv.constants.GOVERNANCE_STATUS_OBSOLETE
pyescv.accept(model)
assert model.status == pyescv.constants.GOVERNANCE_STATUS_ACCEPTED
pyescv.reset(model)
assert model.status == pyescv.constants.GOVERNANCE_STATUS_PENDING
```

```python
institutes = esgf['institute']
pyescv.reject([i for i in institutes if "/" in i.name])
pyescv.reject([i for i in institutes if " " in i.name])
pyescv.reject([i for i in institutes if "," in i.name])
pyescv.accept([institutes['ipsl'], institutes['knmi']])
for i in institutes:
    print i.name, ' --> ', i.status
```
